```python
def f(x, precision):
    x_prec = precision(x)
    val = precision(324.0) * (np.sin(precision(2.5) * x_prec)**2) / (x_prec**2)
    return val

print("i  |  f(x) single       |  f(x) double")
print("-" * 40)

for i in range(11, 21):
    x_value = 10.0**(-i)
    result_single = f(x_value, np.float32)
    result_double = f(x_value, np.float64)
    print(f"{i:2d} |  {result_single:<17.8f} |  {result_double:.8f}")
```

```python
C1 = np.float64(98.0 / 9.0)
C2 = np.float64(11.0 / 9.0)

y0 = np.float64(1.0)
y1 = np.float64(-1.0 / 9.0)

y_values = [y0, y1]

for n in range(48):
    y_next = C1 * y_values[-1] + C2 * y_values[-2]
    y_values.append(y_next)

print(f"{'n':>3} | {'Obliczone y_n':^25} | {'Prawdziwe y_n':^25}")
print("-" * 60)

for n, y_val in enumerate(y_values):
    true_val = (-1.0/9.0)**n
    print(f"{n:3d} | {y_val:^25.15e} | {true_val:^25.15e}")
    if n == 20:
        print("...")
    if n > 45:
        print(f"{n:3d} | {y_val:^25.15e} | {true_val:^25.15e}")
```

```python

num_terms = 2000000

pi_ref = np.pi

# Obliczenia w podwojnen precyzji (dft)
pi_approx_double = 0.0
for k in range(num_terms):
    term = 4.0 * ((-1)**k) / (2*k + 1)
    pi_approx_double += term

# Obliczenia w pojedynczej precyzji
pi_approx_single = np.float32(0.0)
four_single = np.float32(4.0)
one_single = np.float32(1.0)
two_single = np.float32(2.0)
for k in range(num_terms):
    k_single = np.float32(k)
    term = four_single * ((-one_single)**k_single) / (two_single*k_single + one_single)
    pi_approx_single += term


error_double = abs(pi_ref - pi_approx_double)
error_single = abs(np.float32(pi_ref) - pi_approx_single)

print(f"Liczba użytych wyrazów: {num_terms}")
print("-" * 40)
print(f"Wartość referencyjna pi:       {pi_ref:.10f}")
print(f"Przybliżenie (double):          {pi_approx_double:.10f}")
print(f"Błąd (double):                  {error_double:.10f}")
print("-" * 40)
print(f"Przybliżenie (single):          {pi_approx_single:.10f}")
print(f"Błąd (single):                  {error_single:.10f}")
```

```python
import numpy as np

def factorial(n):
    if n == 0:
        return 1
    f = 1
    for i in range(1, n + 1):
        f *= i
    return f

def explicit_cos(x):
    suma = 0.0

    for n in range(100):
        znak = (-1)**n
        licznik = x**(2 * n)
        mianownik = factorial(2 * n)
        wyraz = znak * licznik / mianownik
        poprzednia_suma = suma
        suma += wyraz

        if suma == poprzednia_suma:
            break

    return suma


print(f"{'x (rad)':>10} | {'explicit_cos(x)':^25} | {'np.cos(x)':^25} | {'Błąd bezwzględny':^20}")
print("-" * 88)

test_values = [0, np.pi/6, np.pi/4, np.pi/3, np.pi/2, 2*np.pi/3, np.pi, -np.pi/3]
test_labels = ['0', 'pi/6', 'pi/4', 'pi/3', 'pi/2', '2pi/3', 'pi', '-pi/3']

for label, val in zip(test_labels, test_values):
    my_result = explicit_cos(val)
    lib_result = np.cos(val)
    error = abs(my_result - lib_result)
    print(f"{label:>10} | {my_result:^25.15f} | {lib_result:^25.15f} | {error:^20.2e}")
```

```python
import math

def LOG10(x):
    if 1 <= x <= 10:
        return math.log10(x)

def my_log10(x):
    k = 0

    x_reduced = x
    while x_reduced > 10:
        x_reduced /= 10
        k += 1

    while x_reduced < 1:
        x_reduced *= 10
        k -= 1
    return LOG10(x_reduced) + k


print(f"{'Input x':>12} | {'my_log10(x)':^20} | {'math.log10(x)':^20}")
print("-" * 58)

test_values = [
    5.0,
    4567.8,
    9999.0,
    0.00123,
    0.0001
]

for val in test_values:
    my_result = my_log10(val)
    lib_result = math.log10(val)
    print(f"{val:12.5f} | {my_result:^20.10f} | {lib_result:^20.10f}")
```
