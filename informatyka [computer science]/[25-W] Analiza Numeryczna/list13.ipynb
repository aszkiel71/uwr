```python
import numpy as np
from typing import Callable, Dict, Tuple
import warnings
warnings.filterwarnings("ignore")

def romberg(f: Callable[[np.ndarray], np.ndarray], a: float, b: float, m: int) -> float:
    # romberg table initialization
    t: np.ndarray = np.zeros((m + 1, m + 1))
    h: float = b - a
    t[0, 0] = 0.5 * h * (f(np.array([a])) + f(np.array([b])))

    for i in range(1, m + 1):
        h /= 2
        # trapezoidal update using new midpoints
        sum_new: float = np.sum(f(a + np.arange(1, 2**i, 2) * h))
        t[i, 0] = 0.5 * t[i - 1, 0] + h * sum_new

        # richardson extrapolation
        for k in range(1, i + 1):
            fac: float = 4.0**k
            t[i, k] = (fac * t[i, k - 1] - t[i - 1, k - 1]) / (fac - 1)

    return t[m, m]

# exact values
exact: Dict[str, float] = {
    'a': (2026/9*7**9 + 1977/7*7**7 - 352.5*7**4 - 895.5*7**2 + 7) - \
         (-2026/9 - 1977/7 - 352.5 - 895.5 - 1),
    'b': 0.2 * (np.arctan(5) - np.arctan(-5)),
    'c': 6 * np.log(6) - 5
}

# problem definitions
probs: Dict[str, Tuple[Callable, float, float]] = {
    'a': (lambda x: 2026*x**8 + 1977*x**6 - 1410*x**3 - 1791*x + 1, -1, 7),
    'b': (lambda x: 1 / (1 + 25 * x**2), -1, 1),
    'c': (lambda x: np.log(x + 1), 0, 5)
}

print(f"L13.5 Results (m=20)\n")

res_a = romberg(*probs['a'], 20)
print(f"a) Polynomial:")
print(f"   Romberg: {res_a:.10f}")
print(f"   Exact:   {exact['a']:.10f}")
print(f"   Error:   {abs(res_a - exact['a']):.2e}")

res_b = romberg(*probs['b'], 20)
print(f"\nb) Rational 1/(1+25x^2):")
print(f"   Romberg: {res_b:.15f}")
print(f"   Exact:   {exact['b']:.15f}")
print(f"   Error:   {abs(res_b - exact['b']):.2e}")

res_c = romberg(*probs['c'], 20)
print(f"\nc) Log(x+1):")
print(f"   Romberg: {res_c:.15f}")
print(f"   Exact:   {exact['c']:.15f}")
print(f"   Error:   {abs(res_c - exact['c']):.2e}")
```
