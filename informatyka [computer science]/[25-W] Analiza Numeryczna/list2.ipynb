### 2.1
przeanalizujmy komponenty:

Znak -> dwie możliwości
Mantysa -> ma postać (0.1e[-2]e[-3]e[-4]e[-5]).
Bit e[-1] jest ustawiony na 1.
Mamy 4 zmienne bity (e[-2] do e[-5]) co daje 2^4 = 16 mozliwości.
M = 1/2 + e2/4 + e3/8 + e4/16 + e5/32
Najmniejsza mantysa = (0.10000)2 = 16/32
Nawjeiskza mantysa = (0.11111)2 = 31/32
Zbior mantys: {16/32, 17/32, ..., 31/32}

Wykładnik E ma postać 2^+/-c, gdzie c nalezy do {0, 1}
c = 0 => 2^0 = 1
c = 1 => 2 lub 1/2
Zbior mnoznikow E = {0.5, 1, 2}
2 znaki * 16 mantysy * 3 wykladniki = 96 unikalnych liczb


Zestaw 1 (E = 0.5): x = +/- k/32 * 0.5 = +/- k/64 dla k z przedzialu [16, 31]
32 liczby łącznie
16/64, 17/64, ..., 31/64

Zestaw 2 (E = 1.0): x = +/- k/32 * 1 = dla k z przedzialu [16, 31]
32 liczby lacznie
+/- 32/64, 34/64, ..., 62/64

Zestaw 3 (E = 2.0): x = +/- k/32 * 2 = +/- k/16 = +/- 4k/64 dla k z przedzialu [16, 31]
32 liczby lacznie
+/- 16/16, 17/16,...,31/16

Przedział [A, B] musi zawierać wszystkie wygenerowane liczby.
Wartość maksymalna B: dodatni znak, najwieksza mantysa (k=31, M=31/32), najwiekszy mnoznik (E=2)
B = 31/32 * 2 = 31/16
Wartość minimalna A: ujemny znak, najwieksza mantysa, najwiekszy mnoznik (E=2)
A = -31/16

Rysunek:
nie bede robic rysunku essa.
ale jakby co to widac na nim, że rozkład jest symetrycny wzgledem zera.
W zakresie [1/4, 1/2) Liczby to 16/64, 17/64, ...
odstep wynosi 1/64
W zakresie [1/2, 1) Liczby to 32/64, 34/64, 36/64, ...
odstep wynosi 1/32
W zakresie [1, 31/16] Liczby to 64/64, 68/64, ...
odstep wynosi 1/16

wniosek: Liczby maszynowe nie są rozłożone równomiernie.

```python
# 2.1

import numpy as np

# 1. Zbiór mantys M (16 wartości)
# Mantysa to (16 + bity_e_2_do_e_5) / 32
bity = np.arange(0, 16) # 0000 do 1111
mantysy_licznik = 16 + bity
mantysy = mantysy_licznik / 32.0
# mantysy = [16/32, 17/32, ..., 31/32]

# 2. Zbiór mnożników E (3 wartości)
mnozniki = np.array([0.5, 1.0, 2.0])

# 3. Zbiór znaków S (2 wartości)
znaki = np.array([-1.0, 1.0])

# Generowanie wszystkich 96 liczb
liczby_maszynowe = []
for s in znaki:
    for m in mantysy:
        for e in mnozniki:
            liczby_maszynowe.append(s * m * e)

# Sortowanie i konwersja do tablicy NumPy
liczby_maszynowe = np.sort(np.array(liczby_maszynowe))

print(f"Liczba wygenerowanych liczb: {len(liczby_maszynowe)}")
print(f"Wartość minimalna (A): {liczby_maszynowe[0]}")
print(f"Wartość maksymalna (B): {liczby_maszynowe[-1]}")
```

### 2.2
Standard IEEE to najpowszechniejsza norma techniczna określająca formaty i metody arytmetyczne dla floatow.
Jego celem jest ujednolicenie sposobu reprezentacji i operacji na liczbach rzeczywistych.

x = (-1)^s * M * 2^E
s -> bit znaku
M -> mantysa, część znacząca. znormalizowana tak, że 1 <= M < 2
E -> wykladnik

Liczby subnormalne -> specjalna kategoria liczb w standardzie IEEE, które pozwalaja na reprezentowanie
wartości mniejszych niż najmniejsza dodatnia liczba znormalizowana.
Jak? W przeciwieństwe do liczb znormalizowanych:
1. wykładnik (E) jest ustawiony na swoją minimalną wartość. np. -126 dla float, -1022 dla double
2. Mantysa ma wiodące zero

Różnice:
Wartości specjalne:
-> +/- infty  (nadmiar, overflow) lub wynik operacji typu x/0 (dla x != 0)
reprezentacja: wykładnik = same jedynki, mantysa = same zera
-> NaN, not a number
-> +/- 0, zero z lewej strony, zero z prawej strony
-> Obsluga niedomiaru (underflow) -> jeśli wynik jest mniejszy niż najmniejsza liczba znormalizowana, jest on zbijany do zera. IEEE wprowadza liczby subnormalne, aby zrealizować stopniowy niedomiar.

### 2.6
Algorytm nawiny:
d = x*x
d = d - y*y

kontrprzykład:
x = 1.4 * 1e154
y = 1.3 * 1e154
d = x*x => d = (1.4 * 1e154)^2 = 1.96 * 1e308

Wynik ten przekracza maksymalną wartość dla float64, następuje nadmiar. d <- inf
d = d - y*y = inf - (1.3 * 1e154)^2 = inf - inf = NaN

W rzeczywistości prawdziwa wartość to powinno być 2.7 * 1e307

Proponowany algorytm:
d = (x-y)*(x+y)
Z założenia rd((2*max(|x|, |y|))) należy do Xfl, czyli że nie spowoduje overflow.
dla tych dwóch termów:
|x + y| <= |x| + |y| <= 2*max(|x|, |y|)
|x - y| <= |x| + |y| <= 2*max(|x|, |y|)
Mnożymy dwa czynniki, które nie uległy nadmiarowi.
Wtedy końcowe mnożenie również nie spowoduje nadmiaru.

```python
import numpy as np
```

```python
x = np.float64(1.4e154)
y = np.float64(1.3e154)
```

```python
# algorytm naiwny:
try:
    d1 = x*x
    print(f"Krok 1: (x*x) = {d1}")
    d1 = d1 - (y*y)
    print(f'Krok 2: (x*x) - (y*y) = {d1}')

except Exception as e:
    print(f'Wyjatek: {e}')
```

```python
# algorytm stabilny
try:
    term1 = x - y
    term2 = x + y
    print(f'Krok 1 : (x - y) = {term1}')
    print(f'Krok 2 : (x + y) = {term2}')
    d2 = term1*term2
    print(f'Krok 3 (wynik) = {d2}')
except Exception as e:
    print(f'Wyjatek : {e}')
```

### 2.7
Sprawdźmy, jak komputer (w standardzie double, czyli float64) obliczy tę wartość:

1. obliczanie x^15 :  
        x = 0.01 => (10^-2)^15 = 1e-30
2. obliczanie wnętrza pierwiastka:
        12150*x^15 + 9 = 12150*1e-30 + 9 = 1.215 * 1e-26 + 9.0.
        Wynik w tej arytmertyce ma ok 15-16 cyfr znaczących.
        9.0 + 1.215*1e-26 = 9.000000000000000000000125...
        Arytmetyka double nie jest w stanie przechować tak małej zmiany.
        Róznica między 9.0, a kolejną reprezentowalną liczbą (epsilon) jest rzędu 9.0 * 2e-52 ~ 2*1e-15.
        Dodanie tej wartości "znika" w błędzie zaokrąglenia. Zatem fl(12150*x^15 + 9) = 9.0
3. obliczanie licznika:
        fl(sqrt(9.0)) - 3 = 3.0 - 3.0 = 0.0
4. wynik koncowy:
        Wynik koncowy: fl(f(0.01)) = 0.0/1e-30 = 0.0

Uzasadnienie błędu:
wynik 0.0 jest matematycznie niepoprawny.
Weźmy A = 12150x^15. Ponieważ x = 0.01, to A jest liczbą małą, ale dodatnią. Zatem:
A + 9 > 9
sqrt(A + 9) > sqrt(9) => sqrt(A+9) > 3
Licznik: sqrt(A+9) - 3 > 0
Mianownik: (0.01)e15 > 0

Przyczyną błedu jest 'katastrofalna utrata cyfr znaczących'. (było na wykladzie)
Problem jest w liczniku:
fl(12150x^15 + 9) ~ 9.0.
Cala informacja o małej ale kluczowej wartości 12150x^15 zostaje utracona.
Nastepnie odejmujemy dwie liczby, ktore powinny byc rózne, ale w wyniku bledu zaokraglasnie stały się identyczne.

Proponowany sposob obliczenia:

g(x) = f(x) * 1 = f(x) * (sqrt(12150x^15 + 9) + 3)/(sqrt(12150x^15 + 9) + 3) =
12150/(sqrt(12150x^15 + 9) + 3)

```python
def f(x):
    x = np.float64(x)
    licznik = np.sqrt(12150*np.power(x, 15) + 9) - 3
    mianownik = np.power(x, 15)
    return licznik / mianownik

def g(x):
    x = np.float64(x)
    mianownik = np.sqrt(12150 * np.power(x, 15) + 9) + 3
    return 12150 / mianownik

x_val = 0.01

print(f'f({x_val}) = {f(x_val)}')
print(f'g({x_val}) = {g(x_val)}')
```

### 2.8

f(x) = 162 * (1-cos(5x))/x^2

Problem dotyczy dla bardzo małych x (np. x = 1e-11)
dla x -> 0, mamy 5x -> 0.
Dla x -> 0, cos(x) -> 1.
Problem byłby dla licznika, gdzie mamy 1 - 5cos(x)

1. Obliczanie cos(5x) : gdy x jest bardzo mało np. x = 1e-11, 5x = 5*1e-11.
   Różnica między cos(5*1e-11), a 1.0 jest ekstremalnie mała.
   Z taylora: cos(y) ~ 1 - y^2/2
   cos(5 * 1e-11) ~ 1 - (y^2)/2 = 1 - 1.25 * 1e-21
2. Zaokrąglenie w double : arytmetyka podwójnej precyzji przechowuje ok. 15-16 cyfr znaczących.
   Jej precyzja to ok 2.22 * 1e-16.
   Ponieważ różnica 1.25 * 1e-21 jest znacząco mniejsza niż 2.22 * 1e-16, komputer nie odróżnia
   cos(5*1e-11) od 1.0
   W efekcie fl(cos(5*1e-11)) = 1.0

Propozycja sposobu obliczenia:
cos(2y) = cos^2(y) - sin^2(y) =>
1 - cos(y) = 2sin^2(y/2)
dla y = 5x mamy:
1 - cos(5x) = 2sin^2(5x/2)
f(x) = 162*2sin^2(5x/2)/x^2 = 324*sin^2(2.5x)/x^2

```python
def f(x):
    x = np.float64(x)
    licznik = 1.0 - np.cos(5.0*x)
    mianownik = np.power(x, 2)
    return 162*licznik/mianownik

def g(x):
    x = np.float64(x)
    licznik = 324*np.power(np.sin(2.5*x), 2)
    mianownik = np.power(x, 2)
    return licznik / mianownik

print(f'i  |  f(x)            |   g(x)')

for i in range(1, 21):
    x = 10.0 ** (-i)
    try:
        naive = f(x)
    except Exception:
        naive = np.nan

    try:
        stable = g(x)
    except Exception:
        naive = np.nan
    print(f'{i:2d} | {naive:15.10f}  | {stable:15.10f}')
```

### 2.9

Tak, nastąpi zjawisko utraty cyfr znaczących.
Problem leży w samej formule rekurencyjnej.
Ciąg xk jest zbieżny do pi. W miarę jak k rośnie, xk staje się bliskie pi.
Wartość yk = (xk/2^k)^2 staje się bardzo mała (dla k = 20, yk ~ (pi/2^20)^2 ~ 1e-11, dla k = 3 yk ~ 1e-17).
Kluczowym problem jest w wyrażeniu sqrt(1 - yk). Dla bardzo małego yk, ta wartość jest bardzo bliska 1.
fl(1 - yk) ~ 1.0
1 - 1.0 = 0.0

Zaproponujemy stabilny algoytm.
(na kartce)


```python
# eksperyment

N = 35 # ilosc iteracji

x_unstable = np.zeros(N, dtype=np.float64)
x_stable = np.zeros(N, dtype=np.float64)
z_stable = np.zeros(N, dtype=np.float64)

x_unstable[0] = 2.0  # x_1
z_stable[0] = 1.0    # z_1
x_stable[0] = 2.0**1 * np.sqrt(z_stable[0]) # x_1

for k in range(1, N):

    # k_formula to 'k' ze wzoru x_{k+1} = f(x_k)
    # k_index to indeks tablicy (k_index = k_formula)
    k_formula = k

    # --- Algorytm 1 (Niestabilny) ---
    # x_unstable[k] = x_{k+1}
    # x_unstable[k-1] = x_k
    xk = x_unstable[k-1]
    yk_sq = (xk / (2.0**k_formula))**2

    # Zabezpieczenie przed błędem (gdy pod pierwiastkiem pojawi się NaN)
    if yk_sq > 1.0:
        x_unstable[k] = np.nan
    else:
        inner_sqrt = np.sqrt(1.0 - yk_sq)
        x_unstable[k] = (2.0**k_formula) * np.sqrt(2.0 * (1.0 - inner_sqrt))

    # --- Algorytm 2 (Stabilny) ---
    # z_stable[k] = z_{k+1}
    # z_stable[k-1] = z_k
    zk = z_stable[k-1]

    # Obliczamy z_{k+1}
    if zk > 1.0:
        z_stable[k] = np.nan
    else:
        z_stable[k] = zk / (2.0 * (1.0 + np.sqrt(1.0 - zk)))

    # Obliczamy x_{k+1} = 2^{k+1} * sqrt(z_{k+1})
    # (indeks k to x_{k+1})
    x_stable[k] = (2.0**(k_formula + 1)) * np.sqrt(z_stable[k])


print("--- Testy obliczeniowe dla ciągu x_k -> pi ---")
print(" k |    Algorytm Niestabilny (x_k)    |   Błąd (Niestabilny) |    Algorytm Stabilny (x_k)     |    Błąd (Stabilny)")
print("---|----------------------------------|----------------------|----------------------------------|----------------------")

# Używamy np.pi jako dokładnej wartości
pi_val = np.pi

for k in range(N):
    # k+1 to numer wyrazu ciągu (x_1, x_2, ...)
    idx = k + 1

    val_unstable = x_unstable[k]
    err_unstable = np.abs(pi_val - val_unstable)

    val_stable = x_stable[k]
    err_stable = np.abs(pi_val - val_stable)

    print(f"{idx:2d} | {val_unstable:32.25f} | {err_unstable:20.2e} | {val_stable:32.25f} | {err_stable:20.2e}")
```
