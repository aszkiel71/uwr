```python
import numpy as np
```

```python
# 3.1 a)

def f1(x):
    x = np.float32(x)
    x3 = x**3
    x6 = x**6
    sq = np.sqrt(x6 + np.float32(2025.0))
    res = x3 + sq
    return res

def my_f1(x):
    x = np.float32(x)
    x3 = x**3
    x6 = x**6
    sq = np.sqrt(x6 + np.float32(2025.0))
    if x >= 0:
        res = x3 + sq
    else:
        mianownik = sq - x3
        res = np.float32(2025.0) / mianownik
    return res


x_test_1 = 100.0
print(f"Test dla x = {x_test_1}")
print(f"  Niestabilnie: {f1(x_test_1)}")
print(f"  Stabilnie:     {my_f1(x_test_1)}\n")

x_test_2 = -100.0
print(f"Test dla x = {x_test_2}")
print(f"  Niestabilnie: {f1(x_test_2):.10f}")
print(f"  Stabilnie:     {my_f1(x_test_2):.10f}\n")


x_test_3 = -1000.0
print(f"Test dla x = {x_test_3}")
print(f"  Niestabilnie: {f1(x_test_3):.10f} ")
print(f"  Stabilnie:     {my_f1(x_test_3):.10f} \n")
```

```python
# 3.1 b)
def f2(x):
    x = np.float32(x)
    pi_2 = np.float32(np.pi / 2.0)
    arccot_x = pi_2 - np.arctan(x)
    licznik = pi_2 - x - arccot_x
    res = licznik / (x**3)
    return res

def my_f2(x):
    x = np.float32(x)
    if np.abs(x) < 1e-3:
        x2 = x*x
        x4 = x2*x2
        res = np.float32(-1.0/3.0) + x2 * np.float32(1.0/5.0) - x4 * np.float32(1.0/7.0)
    else:
        licznik = np.arctan(x) - x
        res = licznik / (x**3)
    return res


x_test_4 = 1.0
print(f"Test dla x = {x_test_4} ")
print(f"  Niestabilnie: {f2(x_test_4):.15f}")
print(f"  Stabilnie:     {my_f2(x_test_4):.15f}\n")


x_test_5 = 1e-4
print(f"Test dla x = {x_test_5}:")
print(f"  Niestabilnie: {f2(x_test_5):.15f}  ")
print(f"  Stabilnie:     {my_f2(x_test_5):.15f}  \n")


x_test_6 = 1e-6
print(f"Test dla x = {x_test_6}")
print(f"  Niestabilnie: {f2(x_test_6):.15f}  ")
print(f"  Stabilnie:     {my_f2(x_test_6):.15f}  \n")
```

```python
# 3.2

def retarded(a_in, b_in, c_in):
    a = np.float32(a_in)
    b = np.float32(b_in)
    c = np.float32(c_in)

    delta = b*b - np.float32(4.0)*a*c
    if delta < 0:
        return (np.nan, np.nan)

    sqrt_delta = np.sqrt(delta)

    x1 = (-b + sqrt_delta) / (np.float32(2.0) * a)
    x2 = (-b - sqrt_delta) / (np.float32(2.0) * a)

    return (x1, x2)

def g(a_in, b_in, c_in):
    a = np.float32(a_in)
    b = np.float32(b_in)
    c = np.float32(c_in)

    delta = b*b - np.float32(4.0)*a*c
    if delta < 0:
        return (np.nan, np.nan)

    sqrt_delta = np.sqrt(delta)

    q = -b - np.sign(b) * sqrt_delta

    x_safe = q / (np.float32(2.0) * a)

    x_other = c / (a * x_safe)

    return (x_safe, x_other)


a1, b1, c1 = 1.0, 1e8, 1.0
print(f"a = {a1}, b = {b1}, c = {c1}")
school_roots = retarded(a1, b1, c1)
safe_roots = g(a1, b1, c1)
print(f"school: ({school_roots[0]:.10e}, {school_roots[1]:.10e})")
print(f"stabilne: ({safe_roots[0]:.10e}, {safe_roots[1]:.10e})\n")

a2, b2, c2 = 1.0, -1e8, 1.0
print(f"a = {a2}, b = {b2}, c = {c2}")
school_roots_2 = retarded(a2, b2, c2)
safe_roots_2 = g(a2, b2, c2)
print(f"school: ({school_roots_2[0]:.10e}, {school_roots_2[1]:.10e})")
print(f"stabilne: ({safe_roots_2[0]:.10e}, {safe_roots_2[1]:.10e})\n")

a3, b3, c3 = 1.0, -5.0, 6.0
print(f"a = {a3}, b = {b3}, c = {c3}")
school_roots_3 = retarded(a3, b3, c3)
safe_roots_3 = g(a3, b3, c3)
print(f"school: ({school_roots_3[0]:.10e}, {school_roots_3[1]:.10e})")
print(f"stabilne: ({safe_roots_3[0]:.10e}, {safe_roots_3[1]:.10e})\n")


a4, b4, c4 = 1.0, -5.0, -6.0
print(f"a = {a4}, b = {b4}, c = {c4}")
school_roots_4 = retarded(a4, b4, c4)
safe_roots_4 = g(a4, b4, c4)
print(f"school: ({school_roots_4[0]:.10e}, {school_roots_4[1]:.10e})")
print(f"stabilne: ({safe_roots_4[0]:.10e}, {safe_roots_4[1]:.10e})\n")
```
