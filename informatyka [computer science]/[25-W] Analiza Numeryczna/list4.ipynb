```python
# 4.3

a = 0.0
b = 1.0
b0, a0 = b, a
def f(x):
    x = np.float32(x)
    return x - 0.49

alpha = 0.49
N = 5

res = []

for n in range(0, N + 1):
    m = (a + b) / 2.0
    e = np.abs(alpha - m) # error
    O = 2**(-n-1) * (b0 - a0) # szacowanie
    if n:
        res.append((n, a, b, m, e, O))
    if f(m) > 0:
        b = m
    else:
        a = m


print(f"Zadanie L4.3: Metoda bisekcji")
print(f"Funkcja: f(x) = x - 0.49")
print(f"Przedział początkowy: [a0, b0] = [{a0}, {b0}]")
print(f"Prawdziwy pierwiastek: alpha = {alpha}")
print("=" * 80)
print(" n   a_n             b_n             m_{n+1}         |e_n| (błąd)   O_n (oszacowanie)")
print("-" * 80)
for r in res:
    # n, a_n, b_n, m_{n+1}, |e_n|, O_n
    print(f"{r[0]:2}   {r[1]:<15.10f} {r[2]:<15.10f} {r[3]:<15.10f} {r[4]:<14.10f} {r[5]:<14.10f}")
print("=" * 80)

# en < On
```

```python
# 4.4

def f(x):
    return 3*(x**2) - 5*np.cos(7*x-1)


def bisection(f, a, b, eps = 1e-6):
    assert(f(a) * f(b) < 0)
    f_a = f(a)
    while (b - a)/2.0 > eps:
        m = (a + b) / 2.0

        if f(m) == 0:
            return m

        if f_a * f(m) < 0:
            b = m
        else:
            a = m
            f_a = f(m)

    return (a + b) / 2.0


brackets = [
    (-1.20, -0.75),
    (-0.75, -0.30),
    (-0.30, 0.0),
    (0.14, 0.59),
    (0.59, 1.04),
    (1.04, np.sqrt(5/3))
]

res = []
eps = 1e-6
print(f"Zadanie L4.4: Szukanie zer funkcji f(x) = 3x^2 - 5cos(7x - 1)")
print(f"Metoda: Bisekcja")
print(f"Obszar poszukiwań: [-{np.sqrt(5/3):.5f}, {np.sqrt(5/3):.5f}]")
print(f"Wymagany błąd bezwzględny: {eps}")
print("-" * 60)

for i, (a, b) in enumerate(brackets):
    print(f"Szukanie w przedziale [{a:.4f}, {b:.4f}]...")
    root = bisection(f, a, b, eps)
    if root is not None:
        res.append(root)
        print(f"   Znaleziony pierwiastek {i+1}: {root:.10f}")
        print(f"   Wartość f(pierwiastek): {f(root):.2e}")
    print("-" * 60)

print(f"\nZnaleziono wszystkie {len(res)} zera funkcji.")
```

```python
# 4.5 d)


import numpy as np

def calculate_reciprocal(R, x0):
    """
    Oblicza 1/R za pomocą iteracji Newtona x_n+1 = x_n * (2 - x_n * R).
    Zwraca (wynik, liczba_iteracji).
    """

    # Sprawdzenie warunku zbieżności z zadania (c)
    if not (0 < x0 < 1/R):
        print(f"Błąd: Wartość startowa x0={x0} jest poza przedziałem zbieżności (0, {1/R})")
        return None, 0

    x = x0
    x_prev = x0

    # Używamy standardowej precyzji float64
    machine_epsilon = np.finfo(np.float64).eps

    for i in range(1, 200):

        x_next = x * (2.0 - x * R)

        # Obliczamy błąd względny |x_next - x| / |x_next|
        # Jeśli x_next jest bliskie 0, unikamy dzielenia
        if x_next != 0:
            relative_error = np.abs(x_next - x) / np.abs(x_next)
        else:
            relative_error = np.abs(x_next - x)

        # Warunek stopu: błąd względny jest mniejszy niż epsilon maszynowy
        # lub iteracja przestała się zmieniać
        if relative_error < machine_epsilon or x_next == x:
            return x_next, i

        x = x_next

    return x, i # Zwracamy po przekroczeniu limitu iteracji



test_cases = [
    {"R": 3.0, "x0": 0.1},      # R > 1, x0 "daleko"
    {"R": 3.0, "x0": 0.33},     # R > 1, x0 "blisko"
    {"R": 100.0, "x0": 0.005},  # R >> 1
    {"R": 0.5, "x0": 1.0},      # R < 1
    {"R": 7.0, "x0": 1e-6},    # x0 bardzo blisko 0
    {"R": 7.0, "x0": 1/7.0 - 1e-3} # x0 bardzo blisko 1/R
]

print("Zadanie L4.5(d): Testy eksperymentalne")
print(f"Docelowa precyzja: epsilon maszynowy (ok. {np.finfo(np.float64).eps:.1e})")
print("-" * 70)
print(" R       1/R             x0              Wynik           Iteracji")
print("-" * 70)

for case in test_cases:
    R = case["R"]
    x0 = case["x0"]
    true_val = 1.0 / R

    result, iters = calculate_reciprocal(R, x0)

    if result is not None:
        print(f"{R:<7.1f} {true_val:<15.10f} {x0:<15.10f} {result:<15.10f} {iters:^10}")

# Test poza przedziałem zbieżności
print("-" * 70)
print("Test poza przedziałem zbieżności:")
calculate_reciprocal(3.0, 0.7) # 0.7 > 1/3
calculate_reciprocal(3.0, -0.1) # -0.1 < 0
```

```python
# 4.6
import numpy as np

def calculate_inv_sqrt(a, x0, max_iters=20):
    """
    Oblicza 1/sqrt(a) za pomocą iteracji Newtona:
    x_n+1 = x_n * (1.5 - 0.5 * a * x_n**2)
    """

    x = np.float64(x0)
    a = np.float64(a)

    true_val = 1.0 / np.sqrt(a)

    # Sprawdzenie warunku zbieżności
    if not (0 < x0 < true_val):
        print(f"Ostrzeżenie: x0={x0} jest poza gwarantowanym przedziałem zbieżności (0, {true_val:.2e}).")
        # Metoda może nadal zbiec, jeśli jest w szerszym przedziale |I'(x)|<1

    machine_epsilon = np.finfo(np.float64).eps
    history = []

    for i in range(max_iters):
        x_next = x * (1.5 - 0.5 * a * x**2)
        history.append((i, x, np.abs(x - true_val)))

        # Warunek stopu (precyzja maszynowa)
        if np.abs(x_next - x) < machine_epsilon * np.abs(x_next):
            history.append((i+1, x_next, np.abs(x_next - true_val)))
            return x_next, history, True

        x = x_next

    print(f"Metoda nie zbiegła w {max_iters} iteracjach.")
    return x, history, False

# --- Testy Eksperymentalne ---

test_cases = [
    {"a": 4.0, "x0": 0.1},      # alpha = 0.5. x0 jest w (0, 0.5)
    {"a": 100.0, "x0": 0.05},   # alpha = 0.1. x0 jest w (0, 0.1)
    {"a": 0.01, "x0": 1.0},     # alpha = 10. x0 jest w (0, 10)
    {"a": 9876.0, "x0": 0.001}  # alpha = 0.01. x0 jest w (0, 0.01)
]

print("Zadanie L4.6: Eksperymentalne obliczanie 1/sqrt(a) bez dzielenia")
print("-" * 80)
print(f"Algorytm: x_n+1 = x_n * (1.5 - 0.5 * a * x_n**2)")
print(f"Docelowa precyzja: epsilon maszynowy (ok. {np.finfo(np.float64).eps:.1e})")
print("-" * 80)

for case in test_cases:
    a = case["a"]
    x0 = case["x0"]
    true_val = 1.0 / np.sqrt(a)

    print(f"Test dla a = {a} (alpha = {true_val:.6e}), start z x0 = {x0}")

    result, history, converged = calculate_inv_sqrt(a, x0)

    for step in history:
        print(f"  Iter {step[0]:<2}: x = {step[1]:<22.15e} Błąd: {step[2]:.2e}")

    if converged:
        print(f"Zbieżność osiągnięta w {len(history)-1} iteracjach.")
    print("-" * 80)

# Test dywergencji (x0 za duże)
print("Test dywergencji (x0 poza przedziałem zbieżności):")
print("Test dla a = 4.0 (alpha = 0.5), start z x0 = 2.0 (za duży)")
result, history, converged = calculate_inv_sqrt(4.0, 2.0, max_iters=5)
for step in history:
    print(f"  Iter {step[0]:<2}: x = {step[1]:<22.15e} Błąd: {step[2]:.2e}")
```

```python
import numpy as np

def newton_sqrt_naive(a, x0, tol=1e-15, max_iters=50):
    x = np.float64(x0)
    a = np.float64(a)

    if x <= 0:
        print(f"Błąd: Wartość początkowa x0 musi być > 0.")
        return None, 0

    for i in range(max_iters):
        x_next = 0.5 * (x + a / x)

        if np.abs(x_next - x) < tol * np.abs(x_next):
            return x_next, i + 1

        x = x_next

    return x, max_iters

def newton_sqrt_efficient(a, fixed_x0=1.0, tol=1e-15):
    if a <= 0:
        return None, 0

    a = np.float64(a)

    m, c = np.frexp(a)

    if c % 2 == 1:
        m_target = 2.0 * m
        k = (c - 1) // 2
    else:
        m_target = m
        k = c // 2

    y = np.float64(fixed_x0)

    iters = 0
    for i in range(10):
        iters += 1
        y_next = 0.5 * (y + m_target / y)

        if np.abs(y_next - y) < tol * np.abs(y_next):
            y = y_next
            break
        y = y_next

    result = np.ldexp(y, k)

    return result, iters

a_test = 9.0
true_val_a = np.sqrt(a_test)
print("Zadanie L4.7 (Eksperyment 1): Zbieżność metody dla różnych x0")
print(f"Obliczanie sqrt({a_test}) (Prawdziwa wartość: {true_val_a})")
print("-" * 60)
print(" x0          Wynik           Iteracji")
print("-" * 60)

x0_list = [0.01, 1.0, 10.0, 1000.0, 1e6]
for x0 in x0_list:
    res, iters = newton_sqrt_naive(a_test, x0)
    print(f" {x0:<11.1e} {res:<15.10f} {iters:^10}")

print("\nTest dla x0 <= 0:")
newton_sqrt_naive(a_test, 0.0)

print("\n" + "=" * 60)

print("Zadanie L4.7 (Eksperyment 2): Efektywna metoda z normalizacją")
print(f"Stała wartość startowa dla sqrt(m_target): x0 = 1.0")
print("-" * 60)
print(" a           Wynik (efekt.)  Prawdziwy       Iteracji")
print("-" * 60)

a_list = [0.001, 0.5, 4.0, 9.0, 9876.0, 1e20, 1e-30]
for a_val in a_list:
    res, iters = newton_sqrt_efficient(a_val)
    true_val = np.sqrt(a_val)
    print(f" {a_val:<11.1e} {res:<15.10e} {true_val:<15.10e} {iters:^10}")
```
