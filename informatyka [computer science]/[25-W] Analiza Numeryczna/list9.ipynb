```python
# 9.9

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
```

```python
P = np.array([(39.5, 10.5), (30, 20), (6, 6), (13, -12), (63, -12.5), (18.5, 17.5),
              (48, 63), (7, 25.5), (48.5, 49.5), (9, 19.5), (48.5, 35.5), (59, 32.5), (56, 20.5)])

W = np.array([1, 2, 3, 2.5, 6, 1.5, 5, 1, 2, 1, 3, 5, 1])


t = np.linspace(0, 1, 1000)
n = len(P) - 1
B = np.array([comb(n, i) * t**i * (1-t)**(n-i) for i in range(n + 1)])

WB = B * W[:, None]
curve = (WB.T @ P) / WB.sum(axis=0)[:, None]

plt.figure(figsize=(6, 6))
plt.plot(P[:, 0], P[:, 1], 'k--o', alpha=0.3)
plt.plot(curve[:, 0], curve[:, 1], 'r-', linewidth=2)
plt.axis('equal')
plt.show()

print(B)
```
