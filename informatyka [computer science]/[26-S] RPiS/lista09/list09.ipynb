```python
import pandas as pd
import numpy as np
from scipy.stats import norm
from scipy.stats import t

df = pd.read_csv('z0901.csv', header=None)
data = df.values.flatten()
```

$$\overline{X} = \frac{1}{n} \sum_{i=1}^{n} X_i$$

$$S^2 = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \overline{X})^2$$

```python
# zadanie 1

n = len(data)
X_mean = sum(data) / n
S2 = sum((x - X_mean)**2 for x in data) / (n-1) # / n, zalezy od stopni swobody

print(f'Srednia : {X_mean}')
print(f'Wariancja z probki: {S2}')
```

$$z = \frac{\overline{X} - \mu_0}{\sigma / \sqrt{n}}$$

$$\Phi(z) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{z} e^{-\frac{t^2}{2}} dt$$

```python
# zadanie 2

sigma2 = 8
sigma = np.sqrt(sigma2)

# --- hipoteza nr 1 --- #

m_1 = 1.5
z_1 = (X_mean - m_1) / (sigma / np.sqrt(n))

m_2 = 1.75
z_2 = (X_mean - m_2) / (sigma / np.sqrt(n))

phi_z_1 = norm.cdf(z_1)
phi_z_2 = norm.cdf(z_2)

print('Hipoteza dla μ = 1.5')
print(f"z = {z_1}")
print(f"Phi(z) = {phi_z_1}\n")

print('Hipoteza dla μ = 1.75')
print(f"z = {z_2}")
print(f"Phi(z) = {phi_z_2}\n")
```

$$t = \frac{\overline{X} - \mu_0}{\frac{S}{\sqrt{n}}}$$

$$k = n - 1$$

```python
# zadanie 3

S = np.sqrt(S2)
k = n - 1

m_1 = 1.5
t_1 = (X_mean - m_1) / (S / np.sqrt(n))

m_2 = 1.75
t_2 = (X_mean - m_2) / (S / np.sqrt(n))

t_cdf_1 = t.cdf(t_1, df=k)
t_cdf_2 = t.cdf(t_2, df=k)

print(f"Wartosc k (stopnie swobody) = {k}\n")

print('Hipoteza dla μ = 1.5')
print(f"t = {t_1}")
print(f"Dystrybuanta t(k) = {t_cdf_1}\n")

print('Hipoteza dla μ = 1.75')
print(f"t = {t_2}")
print(f"Dystrybuanta t(k) = {t_cdf_2}\n")
```

$$\hat{\lambda} = \overline{X} = \frac{1}{n} \sum_{i=1}^{n} x_i$$

```python
# zadanie 5

df = pd.read_csv('z0905.csv', header=None)
data = df[0].values
lambda_mle = np.mean(data)
print(f"Estymator λ = {lambda_mle}")
```

$$E(X) = n \cdot p$$
$$\overline{X} = n \cdot \hat{p}$$
$$\hat{p} = \frac{\overline{X}}{n}$$

```python
df = pd.read_csv('z0909.csv', header=None)
data = df[0].values

n_trials = 20
X_mean = np.mean(data)

p_est = X_mean / n_trials

print(f"Estymator p = {p_est}")
```
