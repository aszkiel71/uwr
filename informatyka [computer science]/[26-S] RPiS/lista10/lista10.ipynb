```python
import pandas as pd
import numpy as np
from scipy.stats import norm
from scipy.stats import t
```

```python
df = pd.read_csv("z1008.csv", header=None)
sigma : float = 3.5
mu : int = 2
n : int = len(df)
x_bar = df[0].sum() / n
```

$$Z = \frac{\overline{X} - \mu}{\sigma} \cdot \sqrt{n}$$

```python
z_stat = ((x_bar - mu) / sigma) * np.sqrt(n)
f_z = norm.cdf(z_stat)
```

```python
print(f"Rozmiar próby (n): {n}")
print(f"Średnia z próby (X-bar): {x_bar:.4f}")
print(f"Wartość statystyki Z: {z_stat:.4f}")
print(f"Wartość dystrybuanty F_Z(z): {f_z:.4f}")
```

$$T = \frac{\overline{X}-\mu}{S}\cdot\sqrt{n-1}$$

```python
S = df[0].std(ddof=0)
t_stat = ((x_bar - mu) / S) * np.sqrt(n - 1)
f_t = t.cdf(t_stat, df=n-1)

print(f"T: {t_stat:.4f}")
print(f"F_T(t): {f_t:.4f}")
```
