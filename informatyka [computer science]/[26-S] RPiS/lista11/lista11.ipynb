$$S = \sum_{i=1}^n \frac{(X_i - \mu)^2}{\sigma_0^2}$$

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv('z89.csv', header=None)
data = df[0].values
n = len(data)

# zadanie 8

mu = 2.14
sigma_0 = 4.0
var_0 = sigma_0 ** 2

S = np.sum((data - mu)**2) / var_0

cdf_value = stats.chi2.cdf(S, df=n)

print(f"Liczba obserwacji (n): {n}")
print(f"Wartosc statystyki S: {S:.4f}")
print(f"Wartosc dystrybuanty: {cdf_value:.4f}")
```

```python
# zadanie 9

# Obliczamy srednia z proby (estymator nieznanego mu)
x_bar = np.mean(data)

# Obliczanie statystyki S (z uzyciem sredniej z proby)
S_9 = np.sum((data - x_bar)**2) / var_0

# Liczba stopni swobody wynosi n - 1
df_9 = n - 1

# Obliczanie wartosci dystrybuanty (chi-kwadrat z n-1 stopniami swobody)
cdf_value_9 = stats.chi2.cdf(S_9, df=df_9)

print(f"Srednia z proby (x_bar): {x_bar:.4f}")
print(f"Liczba stopni swobody (df): {df_9}")
print(f"Wartosc statystyki S: {S_9:.4f}")
print(f"Wartosc dystrybuanty: {cdf_value_9:.4f}")
```
